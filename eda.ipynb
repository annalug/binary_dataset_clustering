```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

class FixedDimensionDatasetEmbedder:
    """
    Versão corrigida: Gera vetores de dimensão FIXA para qualquer dataset.
    """
    
    def __init__(self, semantic_method='tfidf', embedding_dim=50, max_stat_features=100):
        """
        Inicializa com dimensões fixas.
        
        Args:
            semantic_method: 'tfidf' (padrão) ou 'spacy'
            embedding_dim: Dimensão fixa para embeddings semânticos
            max_stat_features: Número máximo de features estatísticas
        """
        self.semantic_method = semantic_method
        self.embedding_dim = embedding_dim
        self.max_stat_features = max_stat_features
        
        # Para TF-IDF
        if semantic_method == 'tfidf':
            self.tfidf = TfidfVectorizer(max_features=100, stop_words='english')
            self.svd = TruncatedSVD(n_components=embedding_dim, random_state=42)
            self.tfidf_fitted = False
            self.svd_fitted = False
    
    def create_dataset_embedding(self, df, dataset_name=""):
        """
        Cria um vetor de representação de dimensão FIXA.
        
        Args:
            df: DataFrame com colunas binárias (0/1)
            dataset_name: Nome para logging
            
        Returns:
            np.array: Vetor de representação FIXO do dataset
        """
        print(f"\nProcessando: {dataset_name if dataset_name else 'unnamed'}")
        print(f"  Shape: {df.shape}")
        
        # 1. Componente estatístico (dimensão fixa)
        stats_vector = self._extract_fixed_statistical_features(df)
        print(f"  Estatísticas: {len(stats_vector)} dim")
        
        # 2. Componente semântico (dimensão fixa)
        semantic_vector = self._extract_semantic_features(df.columns.tolist())
        print(f"  Semântica: {len(semantic_vector)} dim")
        
        # 3. Combina ambos (dimensão total fixa)
        dataset_embedding = np.concatenate([stats_vector, semantic_vector])
        print(f"  Total: {len(dataset_embedding)} dimensões")
        
        return dataset_embedding
    
    def _extract_fixed_statistical_features(self, df):
        """
        Extrai características estatísticas com dimensão FIXA.
        """
        # 1. Estatísticas globais (sempre as mesmas)
        global_features = [
            df.shape[0],                    # Número de amostras
            df.shape[1],                    # Número de features
            np.mean(df.values),             # Densidade global
            np.mean(np.mean(df.values, axis=0)),  # Média das médias
            np.mean(np.std(df.values, axis=0)),   # Média dos desvios
            np.median(np.mean(df.values, axis=0)), # Mediana das médias
        ]
        
        # 2. Histograma de densidades (fixo: 10 bins)
        all_values = df.values.flatten()
        hist, _ = np.histogram(all_values, bins=10, range=(0, 1))
        hist = hist / len(all_values)  # Normaliza
        global_features.extend(hist.tolist())
        
        # 3. Estatísticas das TOP N colunas mais comuns
        #    Ordena colunas pela média (proporção de 1s)
        column_means = np.mean(df.values, axis=0)
        top_n = min(self.max_stat_features, len(column_means))
        
        # Pega as top N colunas e adiciona suas estatísticas
        if top_n > 0:
            # Ordena e pega as top N
            top_indices = np.argsort(column_means)[-top_n:]
            
            for idx in top_indices:
                col_data = df.iloc[:, idx].astype(float).values
                mean_val = np.mean(col_data)
                std_val = np.std(col_data)
                
                # Calcula skew e kurtosis apenas se houver variância
                if len(np.unique(col_data)) > 1:
                    skew_val = stats.skew(col_data)
                    kurt_val = stats.kurtosis(col_data)
                else:
                    skew_val = 0.0
                    kurt_val = -3.0
                
                global_features.extend([mean_val, std_val, skew_val, kurt_val])
        
        # 4. Preenche com zeros se necessário para ter dimensão fixa
        total_features = 6 + 10 + (top_n * 4)  # globais + histograma + por coluna
        if total_features < self.max_stat_features * 4 + 16:
            # Adiciona zeros para completar
            padding = (self.max_stat_features * 4 + 16) - total_features
            global_features.extend([0.0] * padding)
        
        return np.array(global_features[:self.max_stat_features * 4 + 16])
    
    def _extract_semantic_features(self, column_names):
        """Cria representação semântica de dimensão fixa."""
        
        # Pré-processa nomes
        processed_names = self._preprocess_column_names(column_names)
        
        if self.semantic_method == 'tfidf':
            # Aplica TF-IDF
            if not self.tfidf_fitted:
                tfidf_matrix = self.tfidf.fit_transform(processed_names)
                self.tfidf_fitted = True
            else:
                tfidf_matrix = self.tfidf.transform(processed_names)
            
            # Reduz dimensionalidade
            if not self.svd_fitted:
                column_embeddings = self.svd.fit_transform(tfidf_matrix)
                self.svd_fitted = True
            else:
                column_embeddings = self.svd.transform(tfidf_matrix)
        else:
            # Fallback simples: one-hot encoding baseado em palavras-chave
            keywords = ['sms', 'location', 'camera', 'audio', 'network', 
                       'storage', 'contact', 'phone', 'internet', 'permission']
            
            column_embeddings = np.zeros((len(column_names), len(keywords)))
            for i, name in enumerate(processed_names):
                for j, keyword in enumerate(keywords):
                    if keyword in name.lower():
                        column_embeddings[i, j] = 1.0
        
        # Agregações (dimensão fixa: 3 * embedding_dim)
        aggregations = []
        
        # 1. Média
        aggregations.append(np.mean(column_embeddings, axis=0))
        
        # 2. Máximo
        aggregations.append(np.max(column_embeddings, axis=0))
        
        # 3. Desvio padrão
        aggregations.append(np.std(column_embeddings, axis=0))
        
        # Concatena todas as agregações
        semantic_vector = np.concatenate(aggregations)
        
        # Garante dimensão correta
        if len(semantic_vector) != self.embedding_dim * 3:
            # Redimensiona se necessário
            from sklearn.decomposition import PCA
            pca = PCA(n_components=self.embedding_dim * 3)
            semantic_vector = pca.fit_transform(semantic_vector.reshape(1, -1)).flatten()
        
        return semantic_vector
    
    def _preprocess_column_names(self, column_names):
        """Pré-processa nomes de colunas."""
        processed = []
        for name in column_names:
            # Converte para minúsculas
            name_lower = str(name).lower()
            
            # Remove prefixos comuns
            prefixes = ['android.permission.', 'dangerous permission:', 
                       'permission:', 'suspicious intent filter:',
                       'com.android.', 'balanced_', '_balanced']
            
            for prefix in prefixes:
                name_lower = name_lower.replace(prefix, '')
            
            # Remove caracteres especiais
            name_lower = name_lower.replace('_', ' ').replace('.', ' ')
            name_lower = ' '.join(name_lower.split())
            
            processed.append(name_lower if name_lower else 'unknown')
        
        return processed

# ============================================================================
# FUNÇÃO PRINCIPAL CORRIGIDA
# ============================================================================

def process_datasets_fixed():
    """Processa todos os datasets com dimensões fixas."""
    
    print("=" * 60)
    print("PROCESSANDO DATASETS COM DIMENSÕES FIXAS")
    print("=" * 60)
    
    # 1. Carregar datasets
    datasets = {
        'adroit': pd.read_csv('data/balanced_adroit.csv'),
        'androcrawl': pd.read_csv('data/balanced_androcrawl.csv'),
        'android_p': pd.read_csv('data/balanced_android_permissions.csv'),
        'drebin': pd.read_csv('data/balanced_drebin215.csv'),
        'defensedroid_close': pd.read_csv('data/balanced_defensedroid_apicalls_closeness.csv'),
        'defensedroid_deg': pd.read_csv('data/balanced_defensedroid_apicalls_degree.csv'),
        'defensedroid_katz': pd.read_csv('data/balanced_defensedroid_apicalls_katz.csv'),
        'defensedroid_prs': pd.read_csv('data/balanced_defensedroid_prs.csv'),
        'kronodroid_emulator': pd.read_csv('data/balanced_kronodroid_emulator.csv'),
        'kronodroid_real': pd.read_csv('data/balanced_kronodroid_real_device.csv')
    }
    
    print(f"Carregados {len(datasets)} datasets")
    
    # 2. Inicializar embedder com dimensões fixas
    embedder = FixedDimensionDatasetEmbedder(
        semantic_method='tfidf',
        embedding_dim=50,
        max_stat_features=20  # Limita features estatísticas
    )
    
    # 3. Processar cada dataset
    dataset_embeddings = {}
    embedding_sizes = []
    
    for name, df in datasets.items():
        # Remove coluna 'class' se existir
        if 'class' in df.columns:
            df = df.drop('class', axis=1)
        
        # Converte todas as colunas para numérico (0/1)
        for col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
        
        # Cria embedding
        embedding = embedder.create_dataset_embedding(df, dataset_name=name)
        dataset_embeddings[name] = embedding
        embedding_sizes.append(len(embedding))
    
    # 4. Verificar se todos têm o mesmo tamanho
    unique_sizes = set(embedding_sizes)
    if len(unique_sizes) == 1:
        print(f"\n✓ Todos os embeddings têm {list(unique_sizes)[0]} dimensões")
    else:
        print(f"\n✗ Tamanhos diferentes encontrados: {unique_sizes}")
        # Força mesmo tamanho pegando o mínimo
        min_size = min(embedding_sizes)
        for name in dataset_embeddings:
            dataset_embeddings[name] = dataset_embeddings[name][:min_size]
        print(f"  Redimensionados para {min_size} dimensões")
    
    return dataset_embeddings, datasets

# ============================================================================
# CLUSTERIZAÇÃO SIMPLIFICADA
# ============================================================================

def simple_clustering(dataset_embeddings):
    """Clusterização simples e robusta."""
    
    from sklearn.cluster import KMeans
    from sklearn.manifold import TSNE
    import matplotlib.pyplot as plt
    
    print("\n" + "=" * 60)
    print("CLUSTERIZAÇÃO")
    print("=" * 60)
    
    # Converter para matriz
    dataset_names = list(dataset_embeddings.keys())
    X = np.array([dataset_embeddings[name] for name in dataset_names])
    
    print(f"Matriz shape: {X.shape}")
    
    # Normalizar
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Redução para visualização (t-SNE)
    print("Reduzindo dimensionalidade para visualização...")
    tsne = TSNE(n_components=2, random_state=42, perplexity=2)
    X_2d = tsne.fit_transform(X_scaled)
    
    # Clusterização com K-means (simples e previsível)
    n_clusters = min(3, len(dataset_names))  # Máximo 3 clusters
    print(f"Aplicando K-means com {n_clusters} clusters...")
    
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    # Visualização
    plt.figure(figsize=(10, 8))
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    
    for i, name in enumerate(dataset_names):
        color = colors[labels[i] % len(colors)]
        plt.scatter(X_2d[i, 0], X_2d[i, 1], color=color, s=200, edgecolors='black')
        plt.text(X_2d[i, 0], X_2d[i, 1], name, fontsize=9, ha='center', va='center')
    
    plt.title('Clusterização de Datasets', fontsize=14)
    plt.xlabel('t-SNE 1')
    plt.ylabel('t-SNE 2')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Mostrar clusters
    print("\n" + "=" * 60)
    print("RESULTADOS")
    print("=" * 60)
    
    clusters = {}
    for name, label in zip(dataset_names, labels):
        clusters.setdefault(label, []).append(name)
    
    for cluster_id in sorted(clusters.keys()):
        members = clusters[cluster_id]
        print(f"\nCluster {cluster_id} ({len(members)} datasets):")
        for member in members:
            print(f"  • {member}")
    
    return labels, X_2d

# ============================================================================
# EXECUÇÃO PRINCIPAL
# ============================================================================

def main_fixed():
    """Execução principal corrigida."""
    
    try:
        # 1. Processar datasets
        dataset_embeddings, datasets = process_datasets_fixed()
        
        # 2. Clusterizar
        labels, coords = simple_clustering(dataset_embeddings)
        
        # 3. Salvar resultados
        import pickle
        
        results = {
            'embeddings': dataset_embeddings,
            'labels': labels,
            'coordinates': coords,
            'dataset_names': list(datasets.keys())
        }
        
        with open('fixed_clustering_results.pkl', 'wb') as f:
            pickle.dump(results, f)
        
        print(f"\n✓ Resultados salvos em 'fixed_clustering_results.pkl'")
        
        # 4. Exemplo de uso com novo dataset
        print("\n" + "=" * 60)
        print("PARA NOVOS DATASETS:")
        print("=" * 60)
        print("""
        # Carregar modelo salvo
        with open('fixed_clustering_results.pkl', 'rb') as f:
            saved = pickle.load(f)
        
        # Criar embedder
        embedder = FixedDimensionDatasetEmbedder()
        
        # Processar novo dataset
        new_df = pd.read_csv('novo_dataset.csv')
        if 'class' in new_df.columns:
            new_df = new_df.drop('class', axis=1)
        
        new_embedding = embedder.create_dataset_embedding(new_df, 'novo')
        
        # Encontrar cluster mais similar
        from sklearn.neighbors import NearestNeighbors
        X = np.array([saved['embeddings'][name] for name in saved['dataset_names']])
        
        nn = NearestNeighbors(n_neighbors=1)
        nn.fit(X)
        dist, idx = nn.kneighbors([new_embedding])
        
        similar_dataset = saved['dataset_names'][idx[0][0]]
        similar_cluster = saved['labels'][idx[0][0]]
        
        print(f"Similar a: {similar_dataset} (Cluster {similar_cluster})")
        """)
        
        return results
        
    except Exception as e:
        print(f"\n✗ Erro: {e}")
        print("\nDebug info:")
        import traceback
        traceback.print_exc()
        
        # Tentar diagnóstico
        diagnose_problem()
        
        return None

def diagnose_problem():
    """Diagnóstico do problema com os datasets."""
    
    print("\n" + "=" * 60)
    print("DIAGNÓSTICO")
    print("=" * 60)
    
    # Verificar um dataset
    try:
        df = pd.read_csv('data/balanced_adroit.csv')
        print(f"Dataset 'adroit' shape: {df.shape}")
        print(f"Colunas: {list(df.columns)[:5]}...")
        print(f"Tipos de dados:")
        for col in df.columns[:5]:
            print(f"  {col}: {df[col].dtype}, valores únicos: {df[col].unique()[:3]}")
        
        # Verificar se tem valores não numéricos
        non_numeric = []
        for col in df.columns:
            if df[col].dtype == 'object':
                non_numeric.append(col)
        
        if non_numeric:
            print(f"\n⚠️ Colunas não numéricas encontradas: {non_numeric[:3]}")
            print("Convertendo para numérico...")
            for col in non_numeric:
                df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
            print("✓ Conversão concluída")
        
    except Exception as e:
        print(f"Erro no diagnóstico: {e}")

# ============================================================================
# VERSÃO SIMPLIFICADA (SE A PRINCIPAL FALHAR)
# ============================================================================

def simple_version():
    """Versão ultra-simplificada que deve funcionar."""
    
    print("=" * 60)
    print("VERSÃO SIMPLIFICADA")
    print("=" * 60)
    
    # Carregar apenas alguns datasets para teste
    datasets_to_test = {
        'adroit': 'data/balanced_adroit.csv',
        'androcrawl': 'data/balanced_androcrawl.csv',
        'android_p': 'data/balanced_android_permissions.csv'
    }
    
    embeddings = {}
    
    for name, path in datasets_to_test.items():
        print(f"\nProcessando {name}...")
        
        try:
            # Carregar
            df = pd.read_csv(path)
            
            # Remover coluna 'class' se existir
            if 'class' in df.columns:
                df = df.drop('class', axis=1)
            
            # Converter tudo para numérico
            for col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
            
            print(f"  Shape final: {df.shape}")
            
            # Criar embedding SIMPLES
            # 1. Densidade global
            density = np.mean(df.values)
            
            # 2. Proporção de colunas com alta densidade (>0.5)
            high_density_cols = np.mean(np.mean(df.values, axis=0) > 0.5)
            
            # 3. Número de colunas
            num_cols = df.shape[1]
            
            # 4. Número de linhas
            num_rows = df.shape[0]
            
            # 5. Embedding semântico simples baseado em palavras-chave
            keywords = ['sms', 'location', 'network', 'storage', 'camera']
            semantic = np.zeros(len(keywords))
            
            for i, col in enumerate(df.columns[:10]):  # Apenas primeiras 10 colunas
                col_lower = str(col).lower()
                for j, kw in enumerate(keywords):
                    if kw in col_lower:
                        semantic[j] += 1
            
            semantic = semantic / len(df.columns[:10])
            
            # Combinar tudo
            embedding = np.concatenate([
                [density, high_density_cols, num_cols, num_rows],
                semantic
            ])
            
            embeddings[name] = embedding
            print(f"  Embedding criado: {len(embedding)} dimensões")
            
        except Exception as e:
            print(f"  ✗ Erro: {e}")
    
    if embeddings:
        print(f"\n✓ Criados {len(embeddings)} embeddings")
        
        # Clusterizar
        from sklearn.cluster import KMeans
        import matplotlib.pyplot as plt
        
        names = list(embeddings.keys())
        X = np.array([embeddings[name] for name in names])
        
        # Normalizar
        from sklearn.preprocessing import StandardScaler
        X_scaled = StandardScaler().fit_transform(X)
        
        # K-means
        kmeans = KMeans(n_clusters=min(2, len(names)), random_state=42)
        labels = kmeans.fit_predict(X_scaled)
        
        # Visualizar (2D usando PCA)
        from sklearn.decomposition import PCA
        pca = PCA(n_components=2)
        X_2d = pca.fit_transform(X_scaled)
        
        plt.figure(figsize=(8, 6))
        for i, name in enumerate(names):
            plt.scatter(X_2d[i, 0], X_2d[i, 1], s=200, label=name)
            plt.text(X_2d[i, 0], X_2d[i, 1], name, fontsize=10)
        
        plt.title('Clusters Simplificados')
        plt.xlabel('PCA 1')
        plt.ylabel('PCA 2')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        
        print("\nClusters:")
        for i, name in enumerate(names):
            print(f"  {name}: Cluster {labels[i]}")
    
    return embeddings

# ============================================================================
# EXECUTAR
# ============================================================================

if __name__ == "__main__":
    
    print("Escolha uma opção:")
    print("1. Versão completa corrigida")
    print("2. Versão simplificada (recomendado primeiro)")
    
    choice = input("\nDigite 1 ou 2: ").strip()
    
    if choice == "1":
        print("\nExecutando versão completa...")
        results = main_fixed()
    else:
        print("\nExecutando versão simplificada...")
        results = simple_version()
    
    if results:
        print("\n✓ Processamento concluído!")
    else:
        print("\n✗ Processamento falhou. Verifique os dados.")
```

```python
"""
PIPELINE AUTOMÁTICO DE CLUSTERIZAÇÃO COM HUGGING FACE
Versão sem interação no terminal - tudo automático
"""
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import pickle
import warnings
import time
warnings.filterwarnings('ignore')

# ============================================================================
# CONFIGURAÇÕES (AJUSTE AQUI)
# ============================================================================

# CONFIGURAÇÕES DO MODELO
HF_MODEL_NAME = "all-MiniLM-L6-v2"  # Modelo mais rápido e eficiente
MAX_STAT_FEATURES = 100            # Features estatísticas máximas
N_CLUSTERS = 3                     # Número de clusters (ajuste conforme necessidade)

# CONFIGURAÇÕES DE VISUALIZAÇÃO
PLOT_COLORS = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']
PLOT_POINT_SIZE = 300
PLOT_FIGSIZE = (16, 8)

# CONFIGURAÇÕES DE ARQUIVOS
INPUT_DATA_PATH = "data/"          # Caminho para seus CSVs
OUTPUT_RESULTS_FILE = "clustering_results_auto.pkl"
OUTPUT_PLOT_FILE = "clusters_auto.png"

# ============================================================================
# 1. CARREGADOR DE EMBEDDINGS (SEM LANGCHAIN - MAIS SIMPLES)
# ============================================================================

class FastHuggingFaceEmbedder:
    """Embedder rápido e simples usando sentence-transformers diretamente."""
    
    def __init__(self, model_name=HF_MODEL_NAME):
        """Carrega modelo do Hugging Face."""
        try:
            from sentence_transformers import SentenceTransformer
            print(f"🔄 Carregando modelo: {model_name}")
            start_time = time.time()
            
            self.model = SentenceTransformer(f'sentence-transformers/{model_name}')
            self.dimensions = self.model.get_sentence_embedding_dimension()
            
            load_time = time.time() - start_time
            print(f"✅ Modelo carregado: {self.dimensions}D ({load_time:.2f}s)")
            
        except ImportError:
            print("❌ sentence-transformers não instalado.")
            print("   Execute: pip install sentence-transformers")
            raise
    
    def encode_batch(self, texts, batch_size=32):
        """Codifica batch de textos."""
        return self.model.encode(
            texts, 
            batch_size=batch_size,
            show_progress_bar=True,
            convert_to_numpy=True
        )

# ============================================================================
# 2. EMBEDDER DE DATASETS
# ============================================================================

class AutoDatasetEmbedder:
    """Cria embeddings automáticos para datasets."""
    
    def __init__(self, model_name=HF_MODEL_NAME):
        self.embedder = FastHuggingFaceEmbedder(model_name)
        self.embedding_dim = self.embedder.dimensions
    
    def create_embedding(self, df, dataset_name=""):
        """Cria embedding para um dataset."""
        print(f"📊 Processando: {dataset_name} ({df.shape})")
        
        # Parte 1: Estatísticas
        stats_vec = self._extract_stats(df)
        
        # Parte 2: Semântica com Hugging Face
        semantic_vec = self._extract_semantics(df.columns.tolist())
        
        # Combina
        embedding = np.concatenate([stats_vec, semantic_vec])
        
        return embedding
    
    def _extract_stats(self, df):
        """Extrai características estatísticas."""
        features = []
        
        # Globais
        features.extend([
            df.shape[0],                    # N amostras
            df.shape[1],                    # N features
            np.mean(df.values),             # Densidade
            np.mean(np.mean(df.values, axis=0)),  # Média das médias
            np.mean(np.std(df.values, axis=0)),   # Média dos desvios
            np.median(np.mean(df.values, axis=0)) # Mediana
        ])
        
        # Histograma (10 bins)
        all_values = df.values.flatten()
        hist, _ = np.histogram(all_values, bins=10, range=(0, 1))
        features.extend((hist / len(all_values)).tolist())
        
        # Estatísticas das colunas mais importantes
        column_means = np.mean(df.values, axis=0)
        n_to_process = min(MAX_STAT_FEATURES // 4, len(column_means))
        
        if n_to_process > 0:
            top_indices = np.argsort(column_means)[-n_to_process:]
            
            for idx in top_indices:
                col_data = df.iloc[:, idx].astype(float).values
                mean_val = np.mean(col_data)
                std_val = np.std(col_data)
                
                if len(np.unique(col_data)) > 1:
                    skew_val = stats.skew(col_data)
                    kurt_val = stats.kurtosis(col_data)
                else:
                    skew_val, kurt_val = 0.0, -3.0
                
                features.extend([mean_val, std_val, skew_val, kurt_val])
        
        # Preenchimento para tamanho fixo
        target_size = 16 + MAX_STAT_FEATURES
        if len(features) < target_size:
            features.extend([0.0] * (target_size - len(features)))
        
        return np.array(features[:target_size])
    
    def _extract_semantics(self, column_names):
        """Extrai significado semântico."""
        # Pré-processa nomes
        processed = [self._clean_column_name(name) for name in column_names]
        
        # Gera embeddings
        embeddings = self.embedder.encode_batch(processed)
        
        # Agregações
        aggregations = [
            np.mean(embeddings, axis=0),    # Média
            np.max(embeddings, axis=0),     # Máximo
            np.std(embeddings, axis=0)      # Desvio
        ]
        
        # Combina
        semantic_vec = np.concatenate(aggregations)
        
        return semantic_vec
    
    def _clean_column_name(self, name):
        """Limpa nome de coluna."""
        name_str = str(name).lower()
        
        # Remove prefixos
        prefixes = [
            'android.permission.', 'dangerous permission:', 
            'permission:', 'suspicious intent filter:',
            'com.android.', 'balanced_', '_balanced'
        ]
        
        for prefix in prefixes:
            name_str = name_str.replace(prefix, '')
        
        # Limpa caracteres
        name_str = name_str.replace('_', ' ').replace('.', ' ')
        name_str = ' '.join(name_str.split())
        
        return f"Android feature: {name_str}" if name_str else "unknown feature"

# ============================================================================
# 3. PIPELINE PRINCIPAL
# ============================================================================

def load_all_datasets():
    """Carrega todos os datasets automaticamente."""
    datasets = {}
    
    # Lista de arquivos (ajuste se necessário)
    file_mapping = {
        'adroit': 'balanced_adroit.csv',
        'androcrawl': 'balanced_androcrawl.csv',
        'android_p': 'balanced_android_permissions.csv',
        'drebin': 'balanced_drebin215.csv',
        'defensedroid_close': 'balanced_defensedroid_apicalls_closeness.csv',
        'defensedroid_deg': 'balanced_defensedroid_apicalls_degree.csv',
        'defensedroid_katz': 'balanced_defensedroid_apicalls_katz.csv',
        'defensedroid_prs': 'balanced_defensedroid_prs.csv',
        'kronodroid_emulator': 'balanced_kronodroid_emulator.csv',
        'kronodroid_real': 'balanced_kronodroid_real_device.csv'
    }
    
    for name, filename in file_mapping.items():
        try:
            df = pd.read_csv(f"{INPUT_DATA_PATH}{filename}")
            
            # Remove coluna 'class' se existir
            if 'class' in df.columns:
                df = df.drop('class', axis=1)
            
            # Converte para numérico
            for col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
            
            datasets[name] = df
            print(f"✓ {name}: {df.shape}")
            
        except Exception as e:
            print(f"✗ Erro ao carregar {name}: {e}")
    
    return datasets

def create_embeddings(datasets):
    """Cria embeddings para todos os datasets."""
    print(f"\n🧠 Criando embeddings com {HF_MODEL_NAME}...")
    
    embedder = AutoDatasetEmbedder()
    embeddings = {}
    
    for name, df in datasets.items():
        emb = embedder.create_embedding(df, name)
        embeddings[name] = emb
    
    # Verifica consistência
    sizes = [len(e) for e in embeddings.values()]
    if len(set(sizes)) != 1:
        min_size = min(sizes)
        for name in embeddings:
            embeddings[name] = embeddings[name][:min_size]
        print(f"⚠️  Ajustado para {min_size} dimensões")
    
    return embeddings

def cluster_datasets(embeddings_dict):
    """Clusteriza os datasets."""
    print("\n🎯 Clusterizando datasets...")
    
    names = list(embeddings_dict.keys())
    X = np.array([embeddings_dict[name] for name in names])
    
    # Normaliza
    X_scaled = StandardScaler().fit_transform(X)
    
    # Reduz para visualização
    print("   Reduzindo dimensionalidade...")
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(4, len(names)-1))
    X_2d = tsne.fit_transform(X_scaled)
    
    # Clusterização
    print("   Aplicando K-means...")
    n_clusters = min(N_CLUSTERS, len(names))
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    
    return names, X_scaled, X_2d, labels

def create_visualization(names, X_2d, labels):
    """Cria visualização colorida dos clusters."""
    print("📈 Gerando visualização...")
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=PLOT_FIGSIZE)
    
    # Gráfico 1: Clusters coloridos
    for cluster_id in np.unique(labels):
        idx = np.where(labels == cluster_id)[0]
        
        color = PLOT_COLORS[cluster_id % len(PLOT_COLORS)]
        
        ax1.scatter(
            X_2d[idx, 0], X_2d[idx, 1],
            color=color,
            s=PLOT_POINT_SIZE,
            edgecolors='black',
            linewidth=2,
            alpha=0.8,
            label=f'Cluster {cluster_id}'
        )
        
        # Labels
        for i in idx:
            ax1.annotate(
                names[i],
                (X_2d[i, 0], X_2d[i, 1]),
                fontsize=9,
                ha='center',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8)
            )
    
    ax1.set_title('Clusters de Datasets', fontsize=14, fontweight='bold')
    ax1.set_xlabel('t-SNE 1')
    ax1.set_ylabel('t-SNE 2')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Gráfico 2: Distribuição dos clusters
    clusters = {}
    for name, label in zip(names, labels):
        clusters.setdefault(label, []).append(name)
    
    cluster_counts = [len(members) for members in clusters.values()]
    cluster_ids = list(clusters.keys())
    
    bars = ax2.bar(
        [f'Cluster {cid}' for cid in cluster_ids],
        cluster_counts,
        color=[PLOT_COLORS[cid % len(PLOT_COLORS)] for cid in cluster_ids]
    )
    
    # Adiciona valores nas barras
    for bar, count in zip(bars, cluster_counts):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                f'{count}', ha='center', va='bottom', fontweight='bold')
    
    ax2.set_title('Distribuição por Cluster', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Número de Datasets')
    ax2.grid(True, alpha=0.3, axis='y')
    
    plt.suptitle('Análise de Clusterização Automática', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig(OUTPUT_PLOT_FILE, dpi=150, bbox_inches='tight')
    plt.show()
    
    print(f"💾 Figura salva: {OUTPUT_PLOT_FILE}")

def save_results(embeddings_dict, names, labels, X_2d):
    """Salva resultados."""
    results = {
        'embeddings': embeddings_dict,
        'dataset_names': names,
        'cluster_labels': labels.tolist(),
        'coordinates_2d': X_2d.tolist(),
        'config': {
            'model': HF_MODEL_NAME,
            'n_clusters': N_CLUSTERS,
            'timestamp': time.strftime("%Y-%m-%d %H:%M:%S")
        }
    }
    
    with open(OUTPUT_RESULTS_FILE, 'wb') as f:
        pickle.dump(results, f)
    
    print(f"💾 Resultados salvos: {OUTPUT_RESULTS_FILE}")
    
    return results

def show_cluster_details(names, labels):
    """Mostra detalhes dos clusters formados."""
    print("\n" + "="*60)
    print("📋 CLUSTERS FORMADOS")
    print("="*60)
    
    clusters = {}
    for name, label in zip(names, labels):
        clusters.setdefault(label, []).append(name)
    
    for cluster_id in sorted(clusters.keys()):
        members = clusters[cluster_id]
        print(f"\n📦 Cluster {cluster_id} ({len(members)} datasets):")
        print("-" * 40)
        for i, member in enumerate(members, 1):
            print(f"  {i:2d}. {member}")

# ============================================================================
# 4. FUNÇÃO PARA NOVOS DATASETS
# ============================================================================

def process_new_dataset(new_df, results_file=OUTPUT_RESULTS_FILE):
    """Processa um novo dataset usando resultados salvos."""
    print("\n🆕 Processando novo dataset...")
    
    # Carrega resultados
    with open(results_file, 'rb') as f:
        saved = pickle.load(f)
    
    # Cria embedder (MESMO modelo)
    embedder = AutoDatasetEmbedder(HF_MODEL_NAME)
    
    # Pré-processa
    if 'class' in new_df.columns:
        new_df = new_df.drop('class', axis=1)
    
    for col in new_df.columns:
        new_df[col] = pd.to_numeric(new_df[col], errors='coerce').fillna(0)
    
    # Cria embedding
    new_emb = embedder.create_embedding(new_df, "novo_dataset")
    
    # Encontra similaridade
    existing_names = saved['dataset_names']
    existing_embs = np.array([saved['embeddings'][name] for name in existing_names])
    
    # Normaliza
    scaler = StandardScaler()
    all_embs = np.vstack([existing_embs, new_emb.reshape(1, -1)])
    all_scaled = scaler.fit_transform(all_embs)
    
    existing_scaled = all_scaled[:-1]
    new_scaled = all_scaled[-1].reshape(1, -1)
    
    # Vizinho mais próximo
    nn = NearestNeighbors(n_neighbors=3, metric='cosine')
    nn.fit(existing_scaled)
    distances, indices = nn.kneighbors(new_scaled)
    
    # Resultados
    most_similar_idx = indices[0][0]
    most_similar_name = existing_names[most_similar_idx]
    assigned_cluster = saved['cluster_labels'][most_similar_idx]
    similarity = 1 - distances[0][0]
    
    print(f"\n🎯 Resultado:")
    print(f"   • Dataset mais similar: {most_similar_name}")
    print(f"   • Cluster atribuído: {assigned_cluster}")
    print(f"   • Similaridade: {similarity:.3f}")
    
    return {
        'assigned_cluster': int(assigned_cluster),
        'most_similar': most_similar_name,
        'similarity_score': float(similarity)
    }

# ============================================================================
# 5. EXECUÇÃO AUTOMÁTICA
# ============================================================================

def main():
    """Executa todo o pipeline automaticamente."""
    print("="*60)
    print("🚀 PIPELINE AUTOMÁTICO DE CLUSTERIZAÇÃO")
    print("="*60)
    
    try:
        # 1. Carregar datasets
        print("\n1. 📁 Carregando datasets...")
        datasets = load_all_datasets()
        
        if not datasets:
            print("❌ Nenhum dataset carregado. Verifique os arquivos.")
            return None
        
        print(f"   ✓ {len(datasets)} datasets carregados")
        
        # 2. Criar embeddings
        print("\n2. 🧠 Criando embeddings...")
        embeddings = create_embeddings(datasets)
        
        # 3. Clusterizar
        print("\n3. 🎯 Clusterizando...")
        names, X_scaled, X_2d, labels = cluster_datasets(embeddings)
        
        # 4. Visualizar
        print("\n4. 📈 Gerando visualização...")
        create_visualization(names, X_2d, labels)
        
        # 5. Salvar resultados
        print("\n5. 💾 Salvando resultados...")
        results = save_results(embeddings, names, labels, X_2d)
        
        # 6. Mostrar detalhes
        show_cluster_details(names, labels)
        
        # 7. Exemplo de uso com novo dataset
        print("\n" + "="*60)
        print("📝 EXEMPLO DE USO COM NOVO DATASET")
        print("="*60)
        print("""
# Para classificar um novo dataset:
new_df = pd.read_csv('seu_novo_dataset.csv')
result = process_new_dataset(new_df)

# Use o cluster retornado para aplicar os pesos:
cluster_id = result['assigned_cluster']
# Aplique os pesos da super-métrica do cluster {cluster_id}
        """)
        
        print(f"\n✅ Pipeline concluído!")
        print(f"   • Arquivo de resultados: {OUTPUT_RESULTS_FILE}")
        print(f"   • Imagem dos clusters: {OUTPUT_PLOT_FILE}")
        
        return results
        
    except Exception as e:
        print(f"\n❌ Erro no pipeline: {e}")
        import traceback
        traceback.print_exc()
        return None

# ============================================================================
# EXECUTAR
# ============================================================================

if __name__ == "__main__":
    # Executa automaticamente
    results = main()
```
